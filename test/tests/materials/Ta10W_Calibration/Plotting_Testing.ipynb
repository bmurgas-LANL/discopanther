```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('method_Discoflux.csv')
e_2 = df['E_2'].to_numpy()
pk2_2 = df['pk2_2'].to_numpy()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(np.abs(e_2),np.abs(pk2_2))
plt.grid()
plt.xlabel('Strain')
plt.xlabel('Stress (MPa)')
plt.show()
```
